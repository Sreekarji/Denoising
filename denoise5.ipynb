````python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
import torchvision
from torchvision import transforms
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim
import math
import os
from datetime import datetime

# ============================================================================
# GPU OPTIMIZED CONFIGURATION
# ============================================================================
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print("="*70)
print("FIXED GPU IMAGE DENOISING WITH DDIM")
print("="*70)
print(f"Device: {device}")
if torch.cuda.is_available():
    print(f"GPU: {torch.cuda.get_device_name(0)}")
    print(f"CUDA Memory: {torch.cuda.get_device_properties(0).total_memory / 1024**3:.1f} GB")
print("="*70)

# Reproducibility
torch.manual_seed(42)
np.random.seed(42)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(42)
    torch.backends.cudnn.benchmark = True

# ============================================================================
# MODEL ARCHITECTURE
# ============================================================================
class TimeEmbedding(nn.Module):
    def __init__(self, dim):
        super().__init__()
        self.dim = dim

    def forward(self, t):
        half_dim = self.dim // 2
        embeddings = math.log(10000) / (half_dim - 1)
        embeddings = torch.exp(torch.arange(half_dim, device=t.device) * -embeddings)
        embeddings = t[:, None] * embeddings[None, :]
        embeddings = torch.cat([embeddings.sin(), embeddings.cos()], dim=-1)
        return embeddings

class ResidualBlock(nn.Module):
    def __init__(self, in_channels, out_channels, time_dim):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels, out_channels, 3, padding=1)
        self.conv2 = nn.Conv2d(out_channels, out_channels, 3, padding=1)
        self.time_mlp = nn.Linear(time_dim, out_channels)
        self.norm1 = nn.GroupNorm(min(8, out_channels//4), out_channels)
        self.norm2 = nn.GroupNorm(min(8, out_channels//4), out_channels)

        if in_channels != out_channels:
            self.shortcut = nn.Conv2d(in_channels, out_channels, 1)
        else:
            self.shortcut = nn.Identity()

    def forward(self, x, time_emb):
        h = self.conv1(x)
        h = self.norm1(h)
        h = F.silu(h)

        time_emb_proj = self.time_mlp(F.silu(time_emb))
        h = h + time_emb_proj[:, :, None, None]

        h = self.conv2(h)
        h = self.norm2(h)
        h = F.silu(h)

        return h + self.shortcut(x)

class ImprovedDenoisingUNet(nn.Module):
    """Improved U-Net with better capacity"""
    def __init__(self, in_channels=3, out_channels=3, time_dim=256):
        super().__init__()

        self.time_mlp = nn.Sequential(
            TimeEmbedding(time_dim),
            nn.Linear(time_dim, time_dim * 2),
            nn.SiLU(),
            nn.Linear(time_dim * 2, time_dim)
        )

        # Better channel dimensions for quality
        self.enc1 = ResidualBlock(in_channels, 64, time_dim)
        self.enc2 = ResidualBlock(64, 128, time_dim)
        self.enc3 = ResidualBlock(128, 256, time_dim)

        self.pool = nn.MaxPool2d(2)

        self.bottleneck = ResidualBlock(256, 256, time_dim)

        self.up3 = nn.ConvTranspose2d(256, 256, 2, stride=2)
        self.dec3 = ResidualBlock(512, 128, time_dim)

        self.up2 = nn.ConvTranspose2d(128, 128, 2, stride=2)
        self.dec2 = ResidualBlock(256, 64, time_dim)

        self.up1 = nn.ConvTranspose2d(64, 64, 2, stride=2)
        self.dec1 = ResidualBlock(128, 64, time_dim)

        self.out = nn.Conv2d(64, out_channels, 1)

    def forward(self, x, t):
        time_emb = self.time_mlp(t)

        e1 = self.enc1(x, time_emb)
        e2 = self.enc2(self.pool(e1), time_emb)
        e3 = self.enc3(self.pool(e2), time_emb)

        b = self.bottleneck(self.pool(e3), time_emb)

        d3 = torch.cat([self.up3(b), e3], dim=1)
        d3 = self.dec3(d3, time_emb)

        d2 = torch.cat([self.up2(d3), e2], dim=1)
        d2 = self.dec2(d2, time_emb)

        d1 = torch.cat([self.up1(d2), e1], dim=1)
        d1 = self.dec1(d1, time_emb)

        return self.out(d1)

# ============================================================================
# DDIM SCHEDULER
# ============================================================================
class DDIMScheduler:
    def __init__(self, num_timesteps=1000, beta_start=0.0001, beta_end=0.02, device='cuda'):
        self.num_timesteps = num_timesteps
        self.device = device

        self.betas = torch.linspace(beta_start, beta_end, num_timesteps, device=device)
        self.alphas = 1.0 - self.betas
        self.alphas_cumprod = torch.cumprod(self.alphas, dim=0)

        self.sqrt_alphas_cumprod = torch.sqrt(self.alphas_cumprod)
        self.sqrt_one_minus_alphas_cumprod = torch.sqrt(1.0 - self.alphas_cumprod)

    def add_noise(self, x_start, t, noise=None):
        if noise is None:
            noise = torch.randn_like(x_start)

        sqrt_alpha_prod = self.sqrt_alphas_cumprod[t][:, None, None, None]
        sqrt_one_minus_alpha_prod = self.sqrt_one_minus_alphas_cumprod[t][:, None, None, None]

        noisy_sample = sqrt_alpha_prod * x_start + sqrt_one_minus_alpha_prod * noise
        return noisy_sample, noise

    @torch.no_grad()
    def denoise(self, model, noisy_image, num_inference_steps=50, eta=0.0):
        """FIXED: Proper denoising with correct scaling"""
        model.eval()

        step_size = self.num_timesteps // num_inference_steps
        timesteps = torch.arange(self.num_timesteps - 1, -1, -step_size, device=self.device)
        timesteps = torch.cat([timesteps, torch.tensor([0], device=self.device)])

        x = noisy_image.clone()

        for i in range(len(timesteps) - 1):
            t_curr = timesteps[i]
            t_next = timesteps[i + 1]

            t_curr_batch = torch.full((x.shape[0],), t_curr, device=self.device, dtype=torch.long)

            predicted_noise = model(x, t_curr_batch)

            alpha_prod_t = self.alphas_cumprod[t_curr]
            alpha_prod_t_next = self.alphas_cumprod[t_next] if t_next >= 0 else torch.tensor(1.0, device=self.device)

            # Predict x_0 - CRITICAL: proper clamping
            pred_x0 = (x - torch.sqrt(1 - alpha_prod_t) * predicted_noise) / torch.sqrt(alpha_prod_t)
            pred_x0 = torch.clamp(pred_x0, 0, 1)

            # DDIM formula
            dir_xt = torch.sqrt(1 - alpha_prod_t_next) * predicted_noise
            x = torch.sqrt(alpha_prod_t_next) * pred_x0 + dir_xt

        return torch.clamp(x, 0, 1)

# ============================================================================
# FIXED TRAINING FUNCTION
# ============================================================================
def train_denoising_model(model, train_loader, scheduler, num_epochs=60,
                         learning_rate=1e-4, save_dir='checkpoints'):
    os.makedirs(save_dir, exist_ok=True)

    # Mixed precision for speed
    scaler = torch.cuda.amp.GradScaler()
    optimizer = torch.optim.AdamW(model.parameters(), lr=learning_rate, weight_decay=0.01)

    # Warmup + Cosine decay
    warmup_epochs = 5
    def lr_lambda(epoch):
        if epoch < warmup_epochs:
            return (epoch + 1) / warmup_epochs
        else:
            progress = (epoch - warmup_epochs) / (num_epochs - warmup_epochs)
            return 0.5 * (1 + math.cos(math.pi * progress))

    lr_scheduler = torch.optim.lr_scheduler.LambdaLR(optimizer, lr_lambda)
    criterion = nn.MSELoss()

    training_losses = []
    best_loss = float('inf')

    print(f"\nðŸš€ Starting GPU training for {num_epochs} epochs...")
    print(f"Learning rate: {learning_rate} (with warmup + cosine decay)")
    print(f"Batch size: {train_loader.batch_size}")
    print(f"Expected time: ~60-90 minutes")
    print("="*70)

    for epoch in range(num_epochs):
        model.train()
        epoch_losses = []

        progress_bar = tqdm(train_loader, desc=f'Epoch {epoch+1}/{num_epochs}')

        for batch_idx, (images, _) in enumerate(progress_bar):
            images = images.to(device)
            batch_size = images.shape[0]

            t = torch.randint(0, scheduler.num_timesteps, (batch_size,),
                            device=device, dtype=torch.long)

            # Mixed precision forward pass
            with torch.cuda.amp.autocast():
                noisy_images, noise = scheduler.add_noise(images, t)
                predicted_noise = model(noisy_images, t)
                loss = criterion(predicted_noise, noise)

            # Backward pass
            optimizer.zero_grad(set_to_none=True)
            scaler.scale(loss).backward()
            scaler.unscale_(optimizer)
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            scaler.step(optimizer)
            scaler.update()

            # Store loss
            epoch_losses.append(loss.item())

            # Update progress
            if batch_idx % 10 == 0:
                avg_recent = np.mean(epoch_losses[-10:])
                progress_bar.set_postfix({
                    'loss': f'{avg_recent:.6f}',
                    'lr': f'{optimizer.param_groups[0]["lr"]:.2e}'
                })

        # Calculate average loss
        avg_loss = np.mean(epoch_losses)
        training_losses.append(avg_loss)
        lr_scheduler.step()

        print(f'Epoch {epoch+1}/{num_epochs} | Loss: {avg_loss:.6f} | LR: {optimizer.param_groups[0]["lr"]:.2e}')

        # Save best model
        if avg_loss < best_loss:
            best_loss = avg_loss
            torch.save({
                'epoch': epoch,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'loss': avg_loss,
            }, os.path.join(save_dir, 'best_model.pt'))
            print(f'âœ“ Best model saved (loss: {avg_loss:.6f})')

        # Save checkpoint every 10 epochs
        if (epoch + 1) % 10 == 0:
            torch.save(model.state_dict(), os.path.join(save_dir, f'model_epoch_{epoch+1}.pt'))
            torch.cuda.empty_cache()
            print(f'âœ“ Checkpoint saved at epoch {epoch+1}')

    print("\n" + "="*70)
    print("âœ… Training completed!")
    print("="*70)

    return training_losses

# ============================================================================
# EVALUATION FUNCTIONS
# ============================================================================
def calculate_denoising_metrics(original, denoised):
    orig_np = original.cpu().numpy().transpose(1, 2, 0)
    denoised_np = denoised.cpu().numpy().transpose(1, 2, 0)

    orig_np = np.clip(orig_np, 0, 1)
    denoised_np = np.clip(denoised_np, 0, 1)

    psnr_value = psnr(orig_np, denoised_np, data_range=1.0)
    ssim_value = ssim(orig_np, denoised_np, data_range=1.0, channel_axis=2)

    return psnr_value, ssim_value

def evaluate_model(model, scheduler, test_loader, noise_levels=[0.2, 0.3, 0.5],
                   num_inference_steps=50):
    model.eval()

    test_images, _ = next(iter(test_loader))
    test_images = test_images.to(device)

    results = {}

    print("\n" + "="*70)
    print("EVALUATION RESULTS")
    print("="*70)

    for noise_level in noise_levels:
        print(f"\nNoise Level Ïƒ = {noise_level}")
        print("-" * 50)

        noise = torch.randn_like(test_images) * noise_level
        noisy_images = torch.clamp(test_images + noise, 0, 1)

        with torch.no_grad():
            denoised_images = scheduler.denoise(model, noisy_images,
                                               num_inference_steps=num_inference_steps)

        psnr_values = []
        ssim_values = []

        for i in range(len(test_images)):
            psnr_val, ssim_val = calculate_denoising_metrics(test_images[i],
                                                            denoised_images[i])
            psnr_values.append(psnr_val)
            ssim_values.append(ssim_val)

        avg_psnr = np.mean(psnr_values)
        avg_ssim = np.mean(ssim_values)

        results[noise_level] = {
            'original': test_images,
            'noisy': noisy_images,
            'denoised': denoised_images,
            'psnr_values': psnr_values,
            'ssim_values': ssim_values,
            'avg_psnr': avg_psnr,
            'avg_ssim': avg_ssim
        }

        print(f"Average PSNR: {avg_psnr:.2f} dB")
        print(f"Average SSIM: {avg_ssim:.4f}")

    print("="*70)

    return results

def visualize_results(results, save_path='denoising_results.png'):
    noise_levels = sorted(results.keys())
    num_samples = 5

    fig, axes = plt.subplots(len(noise_levels) * 3, num_samples,
                            figsize=(15, 4*len(noise_levels)))

    for level_idx, noise_level in enumerate(noise_levels):
        data = results[noise_level]

        for col in range(num_samples):
            row_orig = level_idx * 3
            ax_orig = axes[row_orig, col] if len(noise_levels) > 1 else axes[0, col]
            img_orig = data['original'][col].cpu().permute(1, 2, 0).numpy()
            ax_orig.imshow(np.clip(img_orig, 0, 1))
            ax_orig.axis('off')
            if col == 0:
                ax_orig.set_ylabel('Original', fontsize=10, weight='bold')

            row_noisy = level_idx * 3 + 1
            ax_noisy = axes[row_noisy, col] if len(noise_levels) > 1 else axes[1, col]
            img_noisy = data['noisy'][col].cpu().permute(1, 2, 0).numpy()
            ax_noisy.imshow(np.clip(img_noisy, 0, 1))
            ax_noisy.axis('off')
            if col == 0:
                ax_noisy.set_ylabel(f'Noisy\nÏƒ={noise_level}', fontsize=10, weight='bold')

            row_denoised = level_idx * 3 + 2
            ax_denoised = axes[row_denoised, col] if len(noise_levels) > 1 else axes[2, col]
            img_denoised = data['denoised'][col].cpu().permute(1, 2, 0).numpy()
            ax_denoised.imshow(np.clip(img_denoised, 0, 1))
            ax_denoised.axis('off')
            if col == 0:
                ax_denoised.set_ylabel(
                    f'Denoised\nPSNR:{data["avg_psnr"]:.1f}dB\nSSIM:{data["avg_ssim"]:.3f}',
                    fontsize=9, weight='bold'
                )

    plt.suptitle('GPU-Optimized Image Denoising with DDIM', fontsize=14, weight='bold')
    plt.tight_layout()
    plt.savefig(save_path, dpi=150, bbox_inches='tight')
    plt.show()

    print(f"\nâœ“ Visualization saved to: {save_path}")

# ============================================================================
# MAIN EXECUTION
# ============================================================================
if __name__ == "__main__":
    print("\nðŸ“Š Loading CIFAR-10 dataset...")

    transform = transforms.Compose([
        transforms.Resize(64),
        transforms.ToTensor(),
    ])

    train_dataset = torchvision.datasets.CIFAR10(
        root='./data', train=True, download=True, transform=transform
    )

    test_dataset = torchvision.datasets.CIFAR10(
        root='./data', train=False, download=True, transform=transform
    )

    BATCH_SIZE = 64
    train_loader = DataLoader(
        train_dataset,
        batch_size=BATCH_SIZE,
        shuffle=True,
        num_workers=4,
        pin_memory=True,
        persistent_workers=True
    )

    test_loader = DataLoader(test_dataset, batch_size=8, shuffle=False)

    print(f"âœ“ Training samples: {len(train_dataset)}")
    print(f"âœ“ Test samples: {len(test_dataset)}")
    print(f"âœ“ Batch size: {BATCH_SIZE}")

    # FIXED SETTINGS FOR QUALITY
    NUM_EPOCHS = 60  # More epochs for convergence
    LEARNING_RATE = 1e-4  # Lower for stability
    NUM_TIMESTEPS = 1000
    NUM_INFERENCE_STEPS = 50

    print(f"\nðŸ—ï¸ Initializing model...")
    model = ImprovedDenoisingUNet(in_channels=3, out_channels=3, time_dim=256).to(device)
    scheduler = DDIMScheduler(num_timesteps=NUM_TIMESTEPS, device=device)

    num_params = sum(p.numel() for p in model.parameters())
    print(f"âœ“ Model parameters: {num_params:,}")
    print(f"âœ“ Model size: ~{num_params * 4 / (1024**2):.1f} MB")

    # Train
    print(f"\n{'='*70}")
    print("STARTING TRAINING")
    print(f"{'='*70}")

    training_losses = train_denoising_model(
        model=model,
        train_loader=train_loader,
        scheduler=scheduler,
        num_epochs=NUM_EPOCHS,
        learning_rate=LEARNING_RATE,
        save_dir='checkpoints'
    )

    # Save final model
    torch.save(model.state_dict(), 'denoising_model_final.pt')
    print("\nâœ“ Final model saved to: denoising_model_final.pt")

    # Evaluate
    print(f"\n{'='*70}")
    print("EVALUATING MODEL")
    print(f"{'='*70}")

    evaluation_results = evaluate_model(
        model=model,
        scheduler=scheduler,
        test_loader=test_loader,
        noise_levels=[0.2, 0.3, 0.5],
        num_inference_steps=NUM_INFERENCE_STEPS
    )

    # Visualize
    visualize_results(evaluation_results, save_path='denoising_results_final.png')

    print("\n" + "="*70)
    print("ðŸŽ‰ GPU TRAINING COMPLETE!")
    print("="*70)
    print(f"âœ“ Expected PSNR: 28-32 dB (excellent quality)")
    print(f"âœ“ Expected SSIM: 0.85-0.92 (high similarity)")
    print(f"âœ“ All results saved")
    print("="*70)
````

````python
import os
import shutil
from datetime import datetime
import zipfile

# ============================================================================
# CREATE RESULTS PACKAGE FOR DOWNLOAD
# ============================================================================

def create_results_package():
    """
    Package all training results, models, and visualizations into a
    downloadable zip file
    """

    print("="*70)
    print("PACKAGING RESULTS FOR DOWNLOAD")
    print("="*70)

    # Create results directory
    results_dir = f'denoising_results_{datetime.now().strftime("%Y%m%d_%H%M%S")}'
    os.makedirs(results_dir, exist_ok=True)

    # Files to include
    files_to_copy = {
        'Models': [
            'denoising_model_final.pt',
            'checkpoints/best_model.pt',
        ],
        'Visualizations': [
            'denoising_results_final.png',
            'training_loss_final.png',
            'project_summary_dashboard.png',
        ],
        'Reports': [
            'project_report.txt',
        ]
    }

    # Copy files
    copied_files = []
    missing_files = []

    for category, files in files_to_copy.items():
        print(f"\nðŸ“ {category}:")
        category_dir = os.path.join(results_dir, category)
        os.makedirs(category_dir, exist_ok=True)

        for file in files:
            if os.path.exists(file):
                dest = os.path.join(category_dir, os.path.basename(file))
                shutil.copy2(file, dest)
                file_size = os.path.getsize(file) / (1024**2)
                print(f"  âœ“ {file} ({file_size:.2f} MB)")
                copied_files.append(file)
            else:
                print(f"  âš  {file} (not found)")
                missing_files.append(file)

    # Copy all checkpoints if they exist
    if os.path.exists('checkpoints'):
        checkpoint_dir = os.path.join(results_dir, 'Checkpoints')
        os.makedirs(checkpoint_dir, exist_ok=True)

        print(f"\nðŸ“ Checkpoints:")
        for file in os.listdir('checkpoints'):
            if file.endswith('.pt'):
                src = os.path.join('checkpoints', file)
                dest = os.path.join(checkpoint_dir, file)
                shutil.copy2(src, dest)
                file_size = os.path.getsize(src) / (1024**2)
                print(f"  âœ“ {file} ({file_size:.2f} MB)")

    # Create summary file
    summary_path = os.path.join(results_dir, 'README.txt')
    with open(summary_path, 'w') as f:
        f.write("="*70 + "\n")
        f.write("IMAGE DENOISING WITH DDIM - RESULTS PACKAGE\n")
        f.write("="*70 + "\n")
        f.write(f"Generated: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n\n")

        f.write("CONTENTS:\n")
        f.write("-" * 70 + "\n\n")

        f.write("1. Models/\n")
        f.write("   - denoising_model_final.pt: Final trained model\n")
        f.write("   - best_model.pt: Best model during training\n\n")

        f.write("2. Visualizations/\n")
        f.write("   - denoising_results_final.png: Denoising results visualization\n")
        f.write("   - training_loss_final.png: Training loss curve\n")
        f.write("   - project_summary_dashboard.png: Summary dashboard\n\n")

        f.write("3. Checkpoints/\n")
        f.write("   - model_epoch_*.pt: Checkpoints from training\n\n")

        f.write("4. Reports/\n")
        f.write("   - project_report.txt: Detailed project report\n\n")

        f.write("="*70 + "\n")
        f.write("HOW TO USE:\n")
        f.write("="*70 + "\n\n")

        f.write("To load the model:\n")
        f.write("```python\n")
        f.write("import torch\n")
        f.write("model = ImprovedDenoisingUNet()\n")
        f.write("model.load_state_dict(torch.load('Models/denoising_model_final.pt'))\n")
        f.write("model.eval()\n")
        f.write("```\n\n")

        if copied_files:
            f.write("Files included: " + str(len(copied_files)) + "\n")
        if missing_files:
            f.write("Files not found: " + str(len(missing_files)) + "\n")

    print(f"\nâœ“ README created")

    # Create zip file
    zip_filename = f'{results_dir}.zip'
    print(f"\nðŸ“¦ Creating zip file: {zip_filename}")

    with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for root, dirs, files in os.walk(results_dir):
            for file in files:
                file_path = os.path.join(root, file)
                arcname = os.path.relpath(file_path, os.path.dirname(results_dir))
                zipf.write(file_path, arcname)

    zip_size = os.path.getsize(zip_filename) / (1024**2)
    print(f"âœ“ Zip file created ({zip_size:.2f} MB)")

    # Clean up temporary directory
    shutil.rmtree(results_dir)

    print("\n" + "="*70)
    print("âœ… PACKAGE CREATED SUCCESSFULLY!")
    print("="*70)
    print(f"Download file: {zip_filename}")
    print(f"Size: {zip_size:.2f} MB")
    print("="*70)

    return zip_filename

# ============================================================================
# GOOGLE COLAB DOWNLOAD
# ============================================================================

def download_results_colab():
    """Download results in Google Colab"""
    try:
        from google.colab import files

        zip_file = create_results_package()

        print("\nðŸ”½ Starting download...")
        files.download(zip_file)
        print("âœ“ Download started! Check your browser's download folder.")

    except ImportError:
        print("âŒ Not running in Google Colab!")
        print("The zip file has been created in your current directory.")
        print("You can download it manually.")

# ============================================================================
# JUPYTER/LOCAL DOWNLOAD HELPER
# ============================================================================

def prepare_download_local():
    """Prepare results for local Jupyter or manual download"""

    zip_file = create_results_package()

    print("\n" + "="*70)
    print("DOWNLOAD INSTRUCTIONS")
    print("="*70)
    print(f"\nðŸ“¥ File ready: {zip_file}")
    print(f"\nFor Google Colab:")
    print("  Run: download_results_colab()")
    print(f"\nFor Jupyter Notebook:")
    print(f"  Right-click '{zip_file}' in file browser â†’ Download")
    print(f"\nFor Command Line:")
    print(f"  The file is in your current directory")
    print("="*70)

    return zip_file

# ============================================================================
# INDIVIDUAL FILE DOWNLOADS
# ============================================================================

def download_individual_files_colab():
    """Download files individually in Colab"""
    try:
        from google.colab import files

        print("="*70)
        print("DOWNLOADING FILES INDIVIDUALLY")
        print("="*70)

        download_files = [
            'denoising_model_final.pt',
            'checkpoints/best_model.pt',
            'denoising_results_final.png',
            'project_report.txt'
        ]

        for file in download_files:
            if os.path.exists(file):
                print(f"\nðŸ”½ Downloading: {file}")
                files.download(file)
                print(f"âœ“ {file} downloaded")
            else:
                print(f"âš  {file} not found")

        print("\nâœ… All available files downloaded!")

    except ImportError:
        print("âŒ Not running in Google Colab!")

# ============================================================================
# MAIN EXECUTION
# ============================================================================

if __name__ == "__main__":
    print("\n" + "="*70)
    print("DOWNLOAD OPTIONS")
    print("="*70)
    print("\n1. download_results_colab() - Download zip in Colab")
    print("2. prepare_download_local() - Prepare zip for local download")
    print("3. download_individual_files_colab() - Download files one by one")
    print("\n" + "="*70)

    # Detect environment and auto-download
    try:
        from google.colab import files
        print("\nâœ“ Google Colab detected!")
        print("\nRunning automatic download...")
        download_results_colab()
    except ImportError:
        print("\nâœ“ Local/Jupyter environment detected!")
        print("\nPreparing files...")
        prepare_download_local()
````
